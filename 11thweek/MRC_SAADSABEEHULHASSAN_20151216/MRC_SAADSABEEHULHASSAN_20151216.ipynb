```python
Ts = ([24.542],[1,4,24.542])
t1,y1 = signal.step2(Ts)
plt.plot(t1,y1)
T1 = signal.lti([73.626],[1,7,31.542,73.626])
t,y = signal.step(T1)
plt.plot(t,y)
T2 = signal.lti([245.42],[1,14,64.542,245.42])
t,y = signal.step(T2)
plt.plot(t,y)
T3 = signal.lti([490.84],[1,24,104.542,490.84])
t,y = signal.step(T3)
plt.plot(t,y)
plt.legend(('Ts','T1','T2','T3'))
plt.show()
```

It is clear that T1, T2, T3 have the same complex poles as T(S) but also they have an additional pole and this additional pole plays a significant role in approximation to T(S). when we take the inverse laplace we get an additional negative exponential with this pole. It can be discerned that higher the value of this pole much fast this negative exponential will decay and its responese would be similar to T(S). Thus according to the observation the pole of T3 is the highest and thus its step response is close to that of T(S).

```python
natualfrequency = math.sqrt(361)
```

```python
dampingratio = 16/(2*natualfrequency)
```

```python
peaktime = math.pi/(natualfrequency*math.sqrt(1- dampingratio**2))
```

```python
percentageOS = math.exp(-(math.pi*dampingratio)/ math.sqrt(1 - dampingratio**2)) * 100
```
